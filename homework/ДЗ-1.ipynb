Напишите программу которая определяет
является ли паросочетание устойчивым
• Вход – таблицы предпочтений и паросочетания.
• Выход – ответ: Устойчивое или Неустойчивое.

```python
def is_stable(preferences_group1, preferences_group2, matching):
    """
    preferences_group1: dict, ключи - участники первой группы, значения - списки участников второй группы по предпочтениям
    preferences_group2: dict, ключи - участники второй группы, значения - списки участников первой группы по предпочтениям
    matching: dict, ключи - участники первой группы, значения - их текущие партнеры из второй группы
    """

    current_partners = {v: k for k, v in matching.items()}
    
    for member1, member2 in matching.items():
        pref_list1 = preferences_group1[member1]
        index2_in_1 = pref_list1.index(member2)
        
        preferred_over_current = pref_list1[:index2_in_1]
        for preferred in preferred_over_current:
            current_partner_of_preferred = current_partners[preferred]
            pref_list2 = preferences_group2[preferred]
            
            if pref_list2.index(member1) < pref_list2.index(current_partner_of_preferred):
                return "Неустойчивое"  
    
    return "Устойчивое"
```

```python
group1_preferences = {
    "A": ["X", "Y", "Z"],
    "B": ["Y", "X", "Z"],
    "C": ["X", "Y", "Z"]
}

group2_preferences = {
    "X": ["B", "A", "C"],
    "Y": ["A", "B", "C"],
    "Z": ["A", "B", "C"]
}

matching = {
    "A": "Y",
    "B": "X",
    "C": "Z"
}
```

```python
result = is_stable(group1_preferences, group2_preferences, matching)
result
```
