Решим задание с помощью метода: два указателя
Сложность: O(n + m)
Память: O(1)

```python
def is_subsequence(S, S_prime):
    i = 0  
    j = 0  

    n = len(S)
    m = len(S_prime)

    while i < n and j < m:
        if S[i] == S_prime[j]:
            j += 1
        i += 1

    return j == m
```

```python
S = [
    "купить Amazon",
    "купить Yahoo",
    "купить eBay",
    "купить Yahoo",
    "купить Yahoo",
    "купить Oracle"
]

S_prime = [
    "купить Yahoo",
    "купить eBay",
    "купить Yahoo",
    "купить Oracle"
]

result = is_subsequence(S, S_prime)

print("Является подпоследовательностью:", result)
```

```python
S_prime2 = [
    "купить Yahoo",
    "купить Oracle",
    "купить eBay"
]

print(is_subsequence(S, S_prime2))
```
