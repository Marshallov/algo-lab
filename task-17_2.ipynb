```python
def optimal_sale_order(r):
    n = len(r)

    # сортировка по возрастанию r
    items = sorted(enumerate(r, start=1), key=lambda x: x[1])

    total = 0
    order = []

    for t, (idx, coef) in enumerate(items):
        price = 100 * (coef ** t)
        total += price
        order.append(idx)

    return order, total


# Пример
r = [0.95, 0.80, 0.60, 0.90]

order, profit = optimal_sale_order(r)

print("Коэффициенты:", r)
print("Оптимальный порядок продаж:", order)
print("Максимальная выручка:", round(profit, 2))
print("Максимальная выручка:", round(profit, 2))
```
