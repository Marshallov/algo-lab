# Упражнение 17(2)
Счёт месяцев начинается с 1:
    цена i-го товара при продаже в месяц t:
        100 * r[i]^t

```python
def optimal_sale_order_dp(r):
    n = len(r)

    size = 1 << n
    dp = [0] * size
    parent = [-1] * size

    for mask in range(size - 1, -1, -1):
        sold = bin(mask).count("1")
        month = sold + 1   # первый месяц = 1

        best = -1
        best_item = -1

        for i in range(n):
            if not (mask & (1 << i)):
                new_mask = mask | (1 << i)

                profit = 100 * (r[i] ** month) + dp[new_mask]

                if profit > best:
                    best = profit
                    best_item = i

        if best_item != -1:
            dp[mask] = best
            parent[mask] = best_item

    # восстановление порядка
    order = []
    mask = 0

    while parent[mask] != -1:
        i = parent[mask]
        order.append(i + 1)
        mask |= (1 << i)

    return order, dp[0]



r = [0.01, 0.5, 0.75]

order, profit = optimal_sale_order_dp(r)

print("Коэффициенты:", r)
print("Оптимальный порядок:", order)
print("Максимальная выручка:", round(profit, 2))
```

Решение перебором:

```python
import itertools

def total_profit(order, r):
    profit = 0
    for t, i in enumerate(order, start=1):
        profit += 100 * (r[i] ** t)
    return profit

def find_best_order(r):
    n = len(r)
    best_order = None
    best_profit = -1

    for perm in itertools.permutations(range(n)):
        profit = total_profit(perm, r)
        if profit > best_profit:
            best_profit = profit
            best_order = perm

    return best_order, best_profit

r = [0.5, 0.01, 0.75]

order, profit = find_best_order(r)

print("Коэффициенты:", r)
print("Оптимальный порядок:", order)
print("Максимальная выручка:", round(profit, 2))
```
