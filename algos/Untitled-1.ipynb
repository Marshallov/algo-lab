```python
def solve_truck_maintenance(schedules, n, m):
    chosen = [None] * n  

    def get_schedule(truck_idx, hour):
        if chosen[truck_idx] is None:
            return None
        stop_hour, stop_warehouse = chosen[truck_idx]
        return schedules[truck_idx][hour] if hour < stop_hour else stop_warehouse

    def backtrack(truck_idx):
        if truck_idx == n:
            return True
        for hour in range(m):
            warehouse = schedules[truck_idx][hour]
            if warehouse == 0:
                continue
            chosen[truck_idx] = (hour, warehouse)
            conflict = False
            for h in range(m):
                seen = set()
                for i in range(truck_idx + 1):
                    pos = get_schedule(i, h)
                    if pos is None:
                        conflict = True
                        break
                    if pos > 0:
                        if pos in seen:
                            conflict = True
                            break
                        seen.add(pos)
                if conflict:
                    break
            if not conflict and backtrack(truck_idx + 1):
                return True
            chosen[truck_idx] = None
        return False

    if backtrack(0):
        return [chosen[i][1] for i in range(n)]
    return None

n, m = map(int, input().split())
schedules = [list(map(int, input().split())) for _ in range(n)]
result = solve_truck_maintenance(schedules, n, m)
for warehouse in result:
    print(warehouse)
```
