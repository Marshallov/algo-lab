```python
def min_base_stations(houses):
    houses.sort()
    
    n = len(houses)
    i = 0
    stations = 0
    stations_positions = []
    
    while i < n:
        x = houses[i]
        
        station_pos = x + 4
        stations_positions.append(station_pos)
        
        stations += 1
        
        cover_limit = x + 8
        
        while i < n and houses[i] <= cover_limit:
            i += 1
    
    print("stations =", stations)
    print("positions =", stations_positions)
    
    return stations
```

```python
examples = [
    [1, 2, 3, 10, 11, 12, 20],
    [0, 1, 2, 3, 4, 5],
    [0, 9, 18, 27],
    [1, 5, 6, 7, 12, 13],
    [100]
]

for houses in examples:
    print("\nhouses =", houses)
    min_base_stations(houses)
```
